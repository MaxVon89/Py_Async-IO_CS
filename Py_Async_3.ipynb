## Tasks
<p>Wrapped coroutines, can be awaited, queued, etc.<br>
Can be piled-up, and ran when the event loop has time.</br>
Tasks are futures under the hood.
</p>

```python
import asyncio

async def fetchData(sec, num) -> str:
    print(f"Fetching data #{num} .")
    await asyncio.sleep(sec)
    print(f"Data #{num} found. ")


async def main():
    task1 = asyncio.create_task(fetchData(1, 1))
    task2 = asyncio.create_task(fetchData(2, 2))
    result1 = await task1
    result2 = await task2
    print(f"Task ")
```
