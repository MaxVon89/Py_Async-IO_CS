## Example for generators using many values

```python
import random
from datetime import datetime 
# import mem_profile
```

### First, using Lists

```python
namesList = ['Max', 'Ella', 'Monica', 'Mat', 'Amy']
majorList = ['Engg', 'CS', 'Arts', 'Hist', 'Med']

# def peepList(noPeep) -> str:
def peepList(noPeep):
    result = []
    for i in range(noPeep):
        person = {
            'ID':i,
            'name':random.choice(namesList),
            'major':random.choice(majorList)
        }
        result.append(person)
    return result
    # return person.to_string()

listPeep = peepList(10)
print(listPeep)
```

## Second, using generators

```python
def genPeep(noPeep) -> str:
    
    for i in range(noPeep):
        person = {
            'ID':i,
            'Name':random.choice(namesList),
            'Major':random.choice(majorList)
        }
        yield person
```

```python
t1 = datetime.now()
listPeep4 = genPeep(1000000)
t2 = datetime.now()

print(f"Time taken by List: {t2-t1}")
```
